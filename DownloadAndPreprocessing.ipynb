# Unzip Data

```python
import glob
import zipfile
```

```python
for zip_file in glob.glob('*zip'):
    zip_ref = zipfile.ZipFile(zip_file, 'r')
    zip_ref.extractall('.')
    zip_ref.close()
```

# Resize image to same size
I will chose 256x256 pixels

```python
def resize_image(path, width=256, height=256):
    img = cv2.imread(path)
    array = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    array = cv2.resize(array, (width, height))
    cv2.imwrite(path,array)
```
